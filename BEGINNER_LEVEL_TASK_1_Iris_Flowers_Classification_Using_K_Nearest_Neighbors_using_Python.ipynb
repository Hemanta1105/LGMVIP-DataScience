#**Model Build**

```
#importing libraies required to train and test the model
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.metrics import classification_report,confusion_matrix
```

##Split data set into training set and test set

```
features_columns=list(data_iris.keys()[:-1])
features_columns
```

```
#data spliting
X = data_iris[features_columns]  #features
y = data_iris["Species"]   #taget
```

```
X.head()
```

```
y.head()
```

```
X_train, X_test, y_train, y_test= train_test_split(X,y, random_state= 0, test_size=0.2) # split data into training and test set
```

#Use k-Nearest Neighbor (KNN)

```
knn = KNeighborsClassifier(n_neighbors= 7)
knn.fit(X_train, y_train)
y_pred_knn=knn.predict(X_test)
y_pred_knn
```

##Model Evaluation for KNN

```
print(confusion_matrix(y_test, y_pred_knn))
print(classification_report(y_test,y_pred_knn))
```

```
print("f1_score\t:", 100*(metrics.f1_score(y_test,y_pred_knn,average='macro')))
print("recall_score\t:",100*(metrics.recall_score(y_test,y_pred_knn,average='macro')))
print("precision_score\t:",100*(metrics.precision_score(y_test,y_pred_knn,average='macro')))
print("KNN model accuracy(in %):", metrics.accuracy_score(y_test, y_pred_knn)*100)
```

#Use SVM

```
s = svm.SVC(kernel="linear")
s.fit(X_train,y_train)
y_pred_svm=s.predict(X_test)
y_pred_svm
```

##Model Evaluation for SVM model

```
print(confusion_matrix(y_test, y_pred_svm))
print(classification_report(y_test,y_pred_svm))
```

```
print("f1_score\t:", 100*(metrics.f1_score(y_test,y_pred_knn,average='macro')))
print("recall_score\t:",100*(metrics.recall_score(y_test,y_pred_knn,average='macro')))
print("precision_score\t:",100*(metrics.precision_score(y_test,y_pred_knn,average='macro')))
print("SVM model accuracy(in %):", metrics.accuracy_score(y_test, y_pred_svm)*100)
```
